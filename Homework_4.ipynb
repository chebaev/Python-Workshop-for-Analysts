```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('kc_house_data.csv', sep=',')
df.head()
```

#### 1.2 Изучите стоимости недвижимости

```python
plt.hist(df['price'])
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество');
```

#### 1.3 Изучите распределение квадратуры жилой

```python
sns.histplot(df['sqft_living'], bins=20)
plt.title('Распределение жилой квадратуры')
plt.xlabel('Кв. метры жилые')
plt.ylabel('Количество');
```

#### 1.4 Изучите распределение года постройки

```python
sns.histplot(df['yr_built'], bins=30)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество')
```

## Задача 2

#### 2.1 Изучите распределение домов от наличия вида на набережную

```python
data = df['waterfront'].value_counts()
plt.title('Распределение домов от наличия вида на набережную', color='blue', fontsize=20)
plt.pie(data, autopct='%1.1f%%')
plt.legend(['Нет вида на набережную', 'Есть вид на набережную']);
```

#### 2.2 Изучите распределение этажей домов

```python
data = df['floors'].value_counts()
plt.title('Распределение количество этажей', color='blue', fontsize=20)
plt.pie(data, autopct='%1.1f%%', labels=data.index);
```

#### 2.3 Изучите распределение состояния домов

```python
data = df['condition'].value_counts()
plt.pie(data, autopct='%1.1f%%', labels=data.index);
```

## Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.<br> Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Чем больше жилплощадь дома, тем выше его стоимость

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Чем выше качество конструкции и дизайна дома, тем выше его стоимость

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Оценка конструкции')
plt.title('Распределение цены по оценке конструкции');
```

Наибольшую стоимость имеют дома с оценкой конструкции и дизайна 13

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5,showfliers=False);
plt.xlabel('Цена')
plt.ylabel('Этажи')
plt.title('Распределение цены по этажам');
```

Наибольшую стоимость имеют дома с количеством этажей 2,5

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Оценка вида')
plt.title('Распределение цены по оценкой вида');
```

Наибольшую стоимость имеют дома с оценкой вида 4
